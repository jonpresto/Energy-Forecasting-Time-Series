# Energy Forecasting Time Series

## Motivation
<p>Regression and time series techniques are used to forecast and visualize energy usage for millions of customers for resource allocation.  With the growing focus on dynamic energy usage, getting accurate forecasts for the use and demand is vitally important.  These energy forecasts allow energy companies to better adjust and plan for future energy demand by consumers, which guides policy and resource management.  Forecasting too low leads to buy energy from other locations or rolling blackouts while forecasting too high leads to overuse of resources.  Finding the optimal balance where supply meets demand is essential for an efficient system.</p>

## Brief description of the two datasets
<p>One way to gain accuracy in the forecasts is to understand what factors impact the energy usage of people, which is why the following datasets where chosen:</p>

* <i>hr_temp_20170201-20200131_subset.csv</i> - this dataset contains hourly (variable <b>DATE</b>) temperature data (variable 
<b>HourlyDryBulbTemperature</b> at a weather station near the area of interest.
* <i>hrl_load_metered - 20170201-20200131.csv</i> - This is a dataset contains hourly (variable <b>datetime_beginning_ept</b>) megawatt usage data (variable <b>mw</b>) for the area in Pennsylvania centered around Duquesne. We are using only three years of data because we want to make sure that we look at recent energy patterns that are still applicable to our current customers.

<p>We are assuming that temperature and energy data are probably related.  The hotter or colder it is outside, the more energy could be consumed by residential and commercial buildings to manage indoor temperature.  By combining and cleaning the datasets, we will attempt to explore relationships and eventually build a model to predict future consumption.

## Set time variables and index for better data manipulation

```python
# hour of day, the day of the week, the month, and the year for each observation
import datetime as dt

energy_file['year'] = energy_file["datetime_beginning_ept"].dt.year
energy_file['month'] = energy_file["datetime_beginning_ept"].dt.month
energy_file['day'] = energy_file["datetime_beginning_ept"].dt.day
energy_file['hour'] = energy_file["datetime_beginning_ept"].dt.hour
energy_file['weekday'] = energy_file["datetime_beginning_ept"].dt.weekday
```

```python
energy_file.head()
#energy_file.drop(['hour'], axis=1, inplace=True)
```

```python
weather_file['year'] = weather_file["DATE"].dt.year
weather_file['month'] = weather_file["DATE"].dt.month
weather_file['day'] = weather_file["DATE"].dt.day
weather_file['hour'] = weather_file["DATE"].dt.hour
weather_file['weekday'] = weather_file["DATE"].dt.weekday
```

```python
weather_file.info()
```

```python
weather_file.head()
```

```python
# left join of energy dataset to the temperature dataset
df = energy_file.merge(weather_file, 
                  on=["year", "month", "day", "hour", "weekday"],
                  how='left')
```

```python
keep = ["DATE", "mw", "HourlyDryBulbTemperature", "hour", "weekday", "month", "year"]
```

```python
df = df[keep]
df = df.rename(columns={"DATE": "date", "HourlyDryBulbTemperature": "temp"})
```

```python
df.head()
```

```python
# set date object as the index of our data set. 
# This will make it easier for plotting as well as forecasting later
df = df.set_index(["date"])
```

```python
# creating bool series True for NaN values for "mw" column
bool_series = pd.isnull(df["mw"])
 
# filtering data
# displaying data only with team = NaN
df[bool_series]
```

```python
# creating bool series True for NaN values for "temp" column
bool_series = pd.isnull(df["temp"])
 
# filtering data
# displaying data only with team = NaN
df[bool_series]
```

As shown above, not all the temperature data was recorded by the station.  We will use linear interpolation to fix these missing values in the temperature data.

```python
# to interpolate the missing values
df["temp"] = df["temp"].interpolate(method ='linear')
```

We also should account for the fact that energy usage can vary depending on what time of the day regardless of season.  If we compare usage betweeen 1:00PM vs 1:00AM, we can see that energy usage tends to higher in the middle of the day during business hours, and lower outside business hours. 

Temperature has a quadratic-looking relationship with energy, so we will introduce a new variable, <b>squared temp</b>, to put into the linear regression model.

```python
df_train.info()
```

```python
import statsmodels.api as sm

lm_1 = sm.OLS.from_formula('mw ~ temp', data=df_train).fit()
lm_1.summary()
```

```python
lm_2 = sm.OLS.from_formula('mw ~ temp + temp**2 + C(weekday)', data=df_train).fit()
lm_2.summary()
```

```python
# standardize temp with mean=0 and std=1 since it is good practice to do this
# if we will raise the variable to a power to help prevent multicollinearity problems.

df_train.loc[:, 'temp_norm'] = df_train.loc[:, 'temp']
temp_mean = df_train.loc[:, 'temp'].mean()
temp_std = df_train.loc[:, 'temp'].std()
df_train.loc[:, 'temp_norm'] = (df_train.loc[:, 'temp_norm'] - temp_mean )/temp_std
df_train.loc[:, 'temp_norm_sq'] = df_train.loc[:, 'temp_norm']**2
##df_train.loc[:, 'temp_norm_sq'] = (df_train.loc[:, 'temp_norm_sq'] - df_train.loc[:, 'temp_norm_sq'].mean())/df_train.loc[:, 'temp_norm_sq'].std()

```

```python
lm_3 = sm.OLS.from_formula('mw ~ temp_norm + temp_norm_sq + C(weekday)', data=df_train).fit()
lm_3.summary()
```

```python
lm_4 = sm.OLS.from_formula('mw ~ temp_norm + temp_norm_sq + C(weekday) + C(hour)', data=df_train).fit()
lm_4.summary()
```

```python
lm_5 = sm.OLS.from_formula('mw ~ temp_norm + temp_norm_sq + C(weekday) + C(hour) + C(month)', data=df_train).fit()
lm_5.summary()
```

```python
lm_6 = sm.OLS.from_formula('mw ~ temp_norm + temp_norm_sq + C(weekday) + C(hour) + temp_norm*C(weekday) + temp_norm*C(hour) + C(month) + temp_norm*C(month)',
                           data=df_train).fit()
lm_6.summary()
```

```python
lm_7 = sm.OLS.from_formula('mw ~ temp_norm + temp_norm_sq + C(weekday) + temp_norm*C(weekday) + C(month) + temp_norm*C(month) + temp_norm_sq*C(month) + C(hour) + temp_norm*C(hour) + temp_norm_sq*C(hour)',
                           data=df_train).fit()
lm_7.summary()
```

```python
lm_8 = sm.OLS.from_formula('mw ~ temp_norm + temp_norm_sq + C(weekday) + temp_norm*C(weekday) + C(month) + temp_norm*C(month) + temp_norm_sq*C(month) + C(hour) + temp_norm*C(hour) + temp_norm_sq*C(hour) + C(weekday)*C(hour)',
                           data=df_train).fit()
lm_8.summary()
```

A good evaluation of the model is mean absolute percentage error (MAPE).  MAPE evaluates on average how far off our predictions are in terms of percentages.  

$$ MAPE =  \frac{1}{n} \sum_{i=1..n} \left\lvert{\frac{Y_i - \hat{Y_i}}{Y_i}}\right\rvert * 100  $$

```python
def MAPE(actual, pred):
    n = actual.shape[0]
    result = sum(abs((actual - pred)/actual)*100/n)
    return round(result,6)
```

```python
# Calculate Mean Absolute Percentage Error (MAPE)
# Output MAPE and adjusted R squared for each model

print("\nmodel #   MAPE        R_sq_adj\n")
print("model 1: ", MAPE(df_train['mw'], lm_1.fittedvalues), "  ", lm_1.rsquared_adj)
print("model 2: ", MAPE(df_train['mw'], lm_2.fittedvalues), "  ", lm_2.rsquared_adj)
print("model 3: ", MAPE(df_train['mw'], lm_3.fittedvalues), "  ", lm_3.rsquared_adj)
print("model 4: ", MAPE(df_train['mw'], lm_4.fittedvalues), "  ", lm_4.rsquared_adj)
print("model 5: ", MAPE(df_train['mw'], lm_5.fittedvalues), "  ", lm_5.rsquared_adj)
print("model 6: ", MAPE(df_train['mw'], lm_6.fittedvalues), "  ", lm_6.rsquared_adj)
print("model 7: ", MAPE(df_train['mw'], lm_7.fittedvalues), "  ", lm_7.rsquared_adj)
print("model 8: ", MAPE(df_train['mw'], lm_8.fittedvalues), "  ", lm_8.rsquared_adj)
```

<i>Model 8</i> yields the best model with the lowest MAPE and highest adjusted $R^2$.  Now we will validate the chosen model against the test dataset.

```python
# enrich test data with temp_sq
df_test.loc[:, 'temp_norm'] = df_test.loc[:, 'temp']
df_test.loc[:, 'temp_norm'] = (df_test.loc[:, 'temp_norm'] - temp_mean)/temp_std
df_test.loc[:, 'temp_norm_sq'] = df_test.loc[:, 'temp_norm']**2
```

```python
ypred = lm_8.predict(df_test)
```

```python
# plot lines

fig = plt.gcf()
fig.set_size_inches(15, 6)

x = ypred.index
y1 = ypred #predictions
y2 = df_test['mw'] #actual
plt.plot(x, y1, label = "pred")
plt.plot(x, y2, label = "mw")

plt.xlabel('Date')
plt.ylabel('MW Hours')
plt.title('Prediction vs Actual on Test Dataset')
plt.legend()
plt.show()
```
